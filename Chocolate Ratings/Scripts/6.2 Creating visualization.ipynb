```python
# First, calculate the count of companies by location
company_location_counts = df_cacao['Company_Location'].value_counts()

# Create a mapping from company location to its count
location_count_map = company_location_counts.to_dict()

# Map the counts to the original DataFrame
df_cacao['Company_Location_Count'] = df_cacao['Company_Location'].map(location_count_map)

# Define the bins and the new labels for rarity
bins = [0, 25, 60, 120, 200, float('inf')]
labels = ['extra rare', 'rare', 'somewhat rare', 'somewhat common', 'most common']

# Create a new column that groups the counts into the specified bins with labels for rarity
df_cacao['Location Rarity Classification'] = pd.cut(df_cacao['Company_Location_Count'], bins=bins, labels=labels, right=False)

# Display the first few rows to verify the new column
print(df_cacao[['Company_Location', 'Company_Location_Count', 'Location Rarity Classification']].head())
```

# Corolation Heatmap

```python
#check for types of data
df_cacao.dtypes
```

```python
# Convert the 'Cocoa_Percent' column from percentages to numeric values
df_cacao['Cocoa_Percent'] = df_cacao['Cocoa_Percent'].str.rstrip('%').astype('float') / 100

# Verify the conversion by displaying the updated column
print(df_cacao['Cocoa_Percent'].head())
```

```python
# Select only numeric columns for correlation
numeric_df_cacao = df_cacao.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_df_cacao.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Plot the correlation heatmap using matshow
plt.matshow(correlation_matrix, cmap='coolwarm')

# Add a color bar to represent the correlation values
plt.colorbar()

# Display the plot
plt.show()
```

```python
# Plot the correlation heatmap using matshow
plt.matshow(correlation_matrix, cmap='coolwarm')

# Add a color bar to represent the correlation values
plt.colorbar()

# Display the plot
plt.savefig("out.png")
```

```python
#create labels
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(correlation_matrix.corr(), fignum=f.number) # type of plot
plt.xticks(range(correlation_matrix.shape[1]), correlation_matrix.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(correlation_matrix.shape[1]), correlation_matrix.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```
