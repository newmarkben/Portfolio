# Importing Data

```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
path = r'C:\Users\Oa\Career Foundry\Data'
```

```python
df_ords_prods_merge = pd.read_pickle(os.path.join(path, 'Prepared Data', 'merged_4.7.pkl'))
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.groupby('product_name')
```

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Finding the mean of spenders by loyalty flag
```

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['mean_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.mean)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_order'] < 10, 'type_of_spender'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_order'] >= 10, 'type_of_spender'] = 'High spender'
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge['frequency_median'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.median)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_median'] > 20, 'type_of_frequency'] = 'Frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['frequency_median'] <= 20) & (df_ords_prods_merge['df_ords_prods_merge'] > 10), 'type_of_frequency'] = 'regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_median'] <= 10, 'type_of_frequency'] = 'Non-Frequent customer'
```

```python
df_ords_prods_merge['type_of_frequency'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.to_pickle(os.path.join(path, 'Prepared Data', 'merged_4.8.pkl'))
```
